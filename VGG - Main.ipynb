```python
import os
import numpy as np
import pandas as pd
from PIL import Image
import keras 
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Dropout, InputLayer
from tensorflow.keras.models import Sequential, Model, load_model
```

```python
directory_path = r'D:\LUNA16\classsification'
```

```python
images = []
image_labels = []
image_height = 48
image_width = 48
```

```python
train_df = pd.read_csv('D:\LUNA16\classsification\Train.csv')
```

```python
for index in range(train_df.shape[0]):
        data = train_df.iloc[:, :].values
        np.random.shuffle(data)
    # For Image
        images = data[:, 1:]
    # For Labels
        labels = data[:, 0]
```

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import glob
import PIL
from PIL import Image
import glob
image=[]
for filename in images:
    print(filename)
    for i in filename:
        print(i)
        print(type(i))
        img_array = np.load(i, allow_pickle=True)
        img_array=img_array.reshape(192,192,3)
        image.append(img_array)
```

```python
print(type(labels))
```

```python
print(image[0])
```

```python
img = np.array(image)
image_labels = np.array(labels)
```

```python
print(img[0])
```

```python

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(img, image_labels, test_size=0.2)
```

```python

# Normalising the data
X_train = X_train/255.0
X_val = X_val/255.0
```

```python
from tensorflow.keras.applications import vgg16
```

```python
vgg16_model = vgg16.VGG16(weights = "imagenet", include_top = False, input_shape=(192,192,3))
```

```python
for layer in vgg16_model.layers:
    layer.trainable = False  
```

```python
vgg16_model.summary()
```

```python
x = vgg16_model.output
x = Flatten()(x)
x = Dense(units = 512, activation = "relu")(x)
x = Dropout(rate = 0.25)(x)
x = Dense(units = 256, activation = "relu")(x)
output = Dense(units = 43, activation = "softmax")(x)
```

```python
vgg16_modified_model = Model(inputs = vgg16_model.input, outputs = output)
```

```python
vgg16_modified_model.compile(
    optimizer="Adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
vgg16_modified_model.summary()
```

```python
X_train = np.asarray(X_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.int32)
X_val = np.asarray(X_val).astype(np.float32)
y_val = np.asarray(y_val).astype(np.int32)
```

```python
epochs = 20
history = vgg16_modified_model.fit(X_train, y_train, validation_data=(X_val, y_val),epochs = epochs, batch_size = 32)
```

```python
vgg16_modified_model.save("VGG16new2.h5")
```
